# Business Funding Data in Kazakhstan

Данные Techsalerator о финансировании бизнеса в Казахстане
Данные Techsalerator о финансировании бизнеса в Казахстане содержат подробный и содержательный обзор важной информации для бизнеса, инвесторов и финансовых аналитиков. Этот набор данных обеспечивает тщательный анализ финансовой деятельности в различных секторах Казахстана, детализируя данные, относящиеся к циклам финансирования, источникам инвестиций и ключевым финансовым этапам.

Топ-5 ключевых областей данных
Название компании: Указывает на компанию, получающую финансирование. Эта информация помогает инвесторам определить потенциальные возможности и позволяет аналитикам отслеживать тенденции финансирования в конкретных отраслях.

Сумма финансирования: Показывает общую сумму финансирования, полученного компанией. Понимание этих величин позволяет получить представление о финансовом состоянии и потенциале роста бизнеса, а также о масштабах инвестиционной деятельности.

Этап финансирования: указывает на начальную стадию финансирования, серию А, серию В или более поздние этапы. Это помогает инвесторам оценить зрелость бизнеса и траекторию роста.

Имя инвестора: Содержит подробную информацию об инвесторах или инвестиционных фирмах, участвующих в проекте. Знание инвесторов помогает оценить надежность источника финансирования и их стратегические интересы.

Дата инвестирования: Указывает, когда финансирование было завершено. Сроки инвестирования могут отражать тенденции рынка, доверие инвесторов и потенциальное влияние на будущее компании.

Топ-5 тенденций финансирования в Казахстане
Технологии и стартапы: В технологическом секторе Казахстана, особенно в области финтеха, искусственного интеллекта и разработки программного обеспечения, наблюдается рост инвестиций, стимулирующих инновации и цифровую трансформацию страны.

Энергетический сектор: Являясь одним из крупнейших в мире производителей энергии, Казахстан продолжает привлекать значительные финансовые средства в свой энергетический сектор, особенно в области нефти, газа и возобновляемых источников энергии.

Горнодобывающая промышленность и природные ресурсы: Учитывая богатые минеральные ресурсы Казахстана, включая уран, уголь и драгоценные металлы, значительные инвестиции направляются в горнодобывающие проекты и технологические достижения в области добычи полезных ископаемых.

Инфраструктура и транспорт: Инвестиции в инфраструктурные проекты, включая транспорт, логистику и связь, по-прежнему имеют решающее значение, поскольку Казахстан позиционирует себя как центральный торговый узел между Европой и Азией.

Сельское хозяйство и пищевая промышленность: Благодаря обширным сельскохозяйственным угодьям, финансирование направляется на модернизацию сельскохозяйственного сектора, повышение продовольственной безопасности и поддержку агротехнологических инноваций для повышения производительности.

```python

df_clean = df[['заявленная цена за единицу в валюте', 'оценочная цена за единицу в валюте']].dropna()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df_clean['заявленная цена за единицу в валюте'], bins=30, color='lightseagreen', edgecolor='black', alpha=0.85)
plt.title('Распределение заявленных цен за единицу (FC)', color='darkblue')
plt.xlabel('Заявленная цена (FC)', color='darkgreen')
plt.ylabel('Частота', color='darkgreen')

plt.subplot(1, 2, 2)
plt.hist(df_clean['оценочная цена за единицу в валюте'], bins=30, color='lightcoral', edgecolor='black', alpha=0.85)
plt.title('Распределение оценочных цен за единицу (FC)', color='darkblue')
plt.xlabel('Оценочная цена (FC)', color='darkgreen')
plt.ylabel('Частота', color='darkgreen')

plt.tight_layout()
plt.show()

# Шаг 3: Анализ различий между заявленными и оценочными ценами
df_clean['разница'] = df_clean['заявленная цена за единицу в валюте'] - df_clean['оценочная цена за единицу в валюте']

# Среднее и медианное различие
mean_diff = df_clean['разница'].mean()
median_diff = df_clean['разница'].median()

print(f"Среднее различие между заявленной и оценочной ценой: {mean_diff}")
print(f"Медианное различие между заявленной и оценочной ценой: {median_diff}")

# Построение графика разницы
plt.figure(figsize=(10, 6))
plt.hist(df_clean['разница'], bins=30, color='lightblue', edgecolor='black', alpha=0.85)
plt.title('Распределение различий между заявленной и оценочной ценами', color='darkblue')
plt.xlabel('Разница (заявленная - оценочная цена)', color='darkgreen')
plt.ylabel('Частота', color='darkgreen')
plt.show()
```

**Исходя из графиков и анализа среднего и медианного различия между заявленной и оценочной ценой, можно сделать несколько выводов:
Большая часть товаров имеет как заявленную, так и оценочную стоимость, близкую к нулю. На гистограммах видно, что большинство товаров оценивается на очень низкую стоимость или вообще не имеет значительной цены для учета.
Среднее различие между заявленной и оценочной стоимостью составляет -60.42, что указывает на то, что в среднем заявленная цена на товар ниже оценочной на 60.42 FC. Это может свидетельствовать о том, что товары недооценены в документации или их стоимость умышленно занижается.
Медианное различие между заявленной и оценочной ценой составляет -0.22, что говорит о том, что для большинства товаров разница незначительна. Однако даже небольшие отклонения могут быть показателем различных изменений в стоимости.
На графиках видно, что для некоторых товаров наблюдается значительное отрицательное различие, до -1000 FC. Это указывает на случаи, когда заявленная стоимость товаров заметно ниже оценочной, что может потребовать более глубокого анализа.
Возможные причины различий:
Заниженные заявленные цены. Компании могут снижать заявленные цены на товары, чтобы уменьшить свои налоговые или таможенные обязательства. Это один из самых распространенных случаев, когда стоимость товаров уменьшается, чтобы сократить затраты.
Разрыв между реальной и задекларированной стоимостью. Когда заявленная стоимость товара значительно ниже оценочной, это может говорить о сокрытии реальной стоимости товаров для минимизации финансовых обязательств. Такая практика может иметь финансовые последствия и вызвать вопросы со стороны регулирующих органов.
Влияние обменных курсов или других факторов на оценочную стоимость. Занижение или завышение оценочной стоимости может быть связано с различиями в методах расчета, которые используют таможенные или налоговые органы. Важно обратить внимание на товары с наибольшими отклонениями и проанализировать, с чем они могут быть связаны.
Неравномерная прозрачность в ценообразовании. В некоторых случаях оценочные цены могут быть выше или ниже заявленных из-за различий в подходах к расчету. Это может вызвать дисбаланс между задекларированной и реальной стоимостью.**

```python
# Шаг 1: Группировка данных по странам экспорта и видам транспорта
routes = df.groupby(['страна экспорта', 'вид транспорта']).size().reset_index(name='count')

# Шаг 2: Визуализация топ-10 популярных маршрутов
top_routes = routes.sort_values(by='count', ascending=False).head(10)

# Столбчатая диаграмма для популярных маршрутов
plt.figure(figsize=(10, 6))
plt.barh(top_routes.apply(lambda x: f"{x['страна экспорта']} ({x['вид транспорта']})", axis=1), 
         top_routes['count'], color='lightblue', edgecolor='black', alpha=0.85)
plt.title('Топ-10 популярных торговых маршрутов (страна экспорта и транспорт)', color='darkblue')
plt.xlabel('Количество сделок', color='darkgreen')
plt.ylabel('Маршрут (страна и транспорт)', color='darkgreen')
plt.gca().invert_yaxis()  # Инвертируем ось Y, чтобы топ маршруты шли сверху
plt.show()
```

```python
# 1. Анализ заниженных заявленных цен (разница между заявленной и оценочной стоимостью)
df['разница в цене'] = df['заявленная цена за единицу в валюте'] - df['оценочная цена за единицу в валюте']

# Выявление сделок, где разница отрицательная и значительная
unprofitable_deals = df[df['разница в цене'] < -100]  # Порог в -100 FC, можно скорректировать

# Визуализация невыгодных сделок
plt.figure(figsize=(10, 6))
plt.hist(unprofitable_deals['разница в цене'], bins=20, color='lightcoral', edgecolor='black', alpha=0.85)
plt.title('Распределение невыгодных сделок (отрицательная разница в цене)', color='darkblue')
plt.xlabel('Разница в цене (FC)', color='darkgreen')
plt.ylabel('Количество сделок', color='darkgreen')
plt.show()

# 2. Анализ импорта с завышенной стоимостью (например, стоимость выше 1000 FC)
high_import_costs = df[df['заявленная цена за единицу в валюте'] > 1000]

# Визуализация завышенных цен на импорт
plt.figure(figsize=(10, 6))
plt.hist(high_import_costs['заявленная цена за единицу в валюте'], bins=20, color='lightblue', edgecolor='black', alpha=0.85)
plt.title('Распределение завышенных цен на импорт', color='darkblue')
plt.xlabel('Заявленная цена (FC)', color='darkgreen')
plt.ylabel('Количество сделок', color='darkgreen')
plt.show()

# 3. Анализ неэффективных маршрутов или видов транспорта
inefficient_transport = df.groupby('вид транспорта')['курс валюты'].mean().reset_index()

# Визуализация затрат на разные виды транспорта
plt.figure(figsize=(10, 6))
plt.bar(inefficient_transport['вид транспорта'], inefficient_transport['курс валюты'], 
        color=['lightgreen', 'lightblue', 'lightcoral'], edgecolor='black', alpha=0.85)
plt.title('Средние затраты на транспорт по видам', color='darkblue')
plt.xlabel('Вид транспорта', color='darkgreen')
plt.ylabel('Средняя стоимость транспортировки (FC)', color='darkgreen')
plt.show()

```

```python
# 2. Сгруппировать данные по странам для экспорта и импорта
export_data = df.groupby('страна экспорта')['количество'].sum().reset_index()
import_data = df.groupby('страна происхождения')['количество'].sum().reset_index()

# Переименуем колонки для удобства
export_data.columns = ['страна', 'объем экспорта']
import_data.columns = ['страна', 'объем импорта']

# 3. Объединим данные по экспорту и импорту
trade_data = pd.merge(export_data, import_data, on='страна', how='outer').fillna(0)

# 4. Рассчитаем разницу между экспортом и импортом
trade_data['разница (экспорт - импорт)'] = trade_data['объем экспорта'] - trade_data['объем импорта']

# 5. Визуализация данных

# Диаграмма объема экспорта и импорта по странам
plt.figure(figsize=(14, 8))
plt.bar(trade_data['страна'], trade_data['объем экспорта'], color='lightblue', label='Экспорт', alpha=0.7)
plt.bar(trade_data['страна'], trade_data['объем импорта'], color='lightcoral', label='Импорт', alpha=0.7)
plt.title('Объем экспорта и импорта по странам', color='darkblue')
plt.xlabel('Страна', color='darkgreen')
plt.ylabel('Объем товаров', color='darkgreen')
plt.xticks(rotation=90)
plt.legend()
plt.grid()
plt.show()

# Диаграмма разницы между экспортом и импортом
plt.figure(figsize=(14, 8))
plt.bar(trade_data['страна'], trade_data['разница (экспорт - импорт)'], color='lightseagreen', edgecolor='black', alpha=0.85)
plt.title('Разница между экспортом и импортом по странам', color='darkblue')
plt.xlabel('Страна', color='darkgreen')
plt.ylabel('Разница (Экспорт - Импорт)', color='darkgreen')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

Объем экспорта и импорта по странам:

Бельгия показала наибольший объем экспорта, превосходя все другие страны.
Нидерланды и Малайзия являются крупнейшими импортерами, причем их объемы импорта значительно превышают объемы экспорта.
Франция и Китай также участвуют в импорте, однако их объемы меньше по сравнению с Нидерландами и Малайзией.
Великобритания и Сингапур демонстрируют незначительные объемы экспорта и импорта.
Разница между экспортом и импортом:

Бельгия выделяется значительным положительным балансом (экспорт значительно превышает импорт).
Нидерланды и Малайзия, напротив, демонстрируют крупный отрицательный баланс (импорт намного больше экспорта).
У Франции и Китая баланс близок к нулю, что говорит о более сбалансированных объемах импорта и экспорта.
